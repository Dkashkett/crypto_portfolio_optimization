```python
import pandas as pd 
import pandas_datareader.data as web
import datetime
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2021,1,1)
```

```python
other = {'Dogecoin':'DOGE-USD', 'Polkadot':'DOT1-USD'}
```

```python
tickers = {'Bitcoin':'BTC-USD', 'Bitcoin Cash':'BCH-USD',
                 'Ethereum':'ETH-USD','Litecoin':'LTC-USD',
                 'Tron':'TRX-USD',
                 'Binance Coin':'BNB-USD',
                 'Ripple':'XRP-USD','Cardano':'ADA-USD',
                 'Stellar':'XLM-USD','Monero':'XMR-USD',
                  'Chainlink':'LINK-USD','Dogecoin':'DOGE-USD', 
                   'Polkadot':'DOT1-USD'}
```

```python
crypto = web.DataReader(list(tickers.values())[0], 'yahoo', start, end).reset_index()[['Date','Close']]
for ticker in list(tickers.values())[1:]:
    frame = web.DataReader(ticker, 'yahoo', start, end).reset_index()[['Date','Close']]
    crypto = crypto.merge(frame, left_on='Date', right_on='Date', how='outer')
crypto = crypto.set_index('Date')
crypto.columns = list(tickers.keys())
```

```python
returns = crypto/crypto.shift(1)
```

```python
log_returns = np.log(returns)
```

```python
n_currencies = len(tickers)
n_portfolios = 10000

weights = np.zeros((n_portfolios,n_currencies))
expected_return = np.zeros(n_portfolios)
expected_volatility = np.zeros(n_portfolios)
sharpe_ratio = np.zeros(n_portfolios)

mean_log_return = log_returns.mean()
sigma = log_returns.cov()

for k in range(n_portfolios):
    w = np.array(np.random.random(n_currencies))
    w = w/np.sum(w)
    weights[k,:] = w
    # expected log returns
    expected_return[k] = np.sum(mean_log_return * w)
    # expected volatility
    expected_volatility[k] = np.sqrt(np.dot(w.T, np.dot(sigma,w)))
    #sharpe ratio
    sharpe_ratio[k] = expected_return[k]/expected_volatility[k] 
    
```

```python
max_index = sharpe_ratio.argmax()
weights[max_index, :]
```

```python
plt.figure(figsize=(10,8))
plt.scatter(expected_volatility, expected_return, c=sharpe_ratio)
plt.scatter(expected_volatility[max_index], 
            expected_return[max_index], c='red', label='Max SR')
plt.xlabel('Expected Volatility')
plt.ylabel('Log Expected Returns')
plt.colorbar(label='Sharpe Ratio')
plt.legend()
```

```python
def negative_sharpe(w):
    w = np.array(w)
    r = np.sum(mean_log_return * w) 
    v = np.sqrt(np.dot(w.T, np.dot(sigma,w)))
    sr = r/v
    return -1*sr

def check_sum_to_one(w):
    return np.sum(w) - 1
```

```python
w0 = [1/n_currencies for i in range(n_currencies)]
bounds = tuple([(0,1) for i in range(n_currencies)])
constraints = ({'type':'eq', 'fun': check_sum_to_one})
w_opt = minimize(negative_sharpe, w0, method ='SLSQP', bounds=bounds,
                constraints=constraints)
```

```python
w_opt.x
```

```python
returns = np.linspace(expected_return.min(),expected_return.max(),100)
volatility_opt = []
def minimize_volatility(w):
    w = np.array(w)
    v = np.sqrt(np.dot(w.T, np.dot(sigma,w)))
    return v

def get_return(w):
    w = np.array(w)
    r = np.sum(mean_log_return*w)
    return r

for r in returns:
    constraints = ({'type':'eq','fun': check_sum_to_one},
                  {'type':'eq','fun':lambda w: get_return(w) - r})
    opt = minimize(minimize_volatility, w0, 
                    method ='SLSQP', bounds=bounds,
                constraints=constraints)
    volatility_opt.append(opt['fun'])
    
```

```python
plt.figure(figsize=(10,8))
plt.scatter(expected_volatility, expected_return, c=sharpe_ratio)
plt.scatter(expected_volatility[max_index], 
            expected_return[max_index], c='red', label='Max SR')
plt.xlabel('Expected Volatility')
plt.ylabel('Log Expected Returns')
plt.colorbar(label='Sharpe Ratio')
plt.plot(volatility_opt, returns,'--')
plt.legend()
```
