```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, newton, show_options, line_search
from scipy import misc
```

```python
def plot_results(results,title='Plot'):
    seq = [f(i)[0] for i in results.allvecs]
    plt.figure(figsize=(8,6))
    plt.plot(list(range(len(seq))), seq)
    plt.title(title)
    plt.ylabel('F(x)')
    plt.xlabel('Iteration #')
    print('Best X: ', results.x)
    return np.array((list(range(len(seq))), seq))
```

# HOMEWORK

### Objective function and x0

```python
x0 = np.array([2,4,6]).reshape((3,1))

def f(x):
    P = np.array([5,4,6,4,9,15,6,15,35]).reshape((3,3))
    q = np.array([5,7,6]).reshape((3,1))
    fofx = 0.5 * (x.T@P)@x + q.T@x
    return fofx

def obj_grad(x):
    return P@x + q
```

### Newton 

```python
P = np.array([5,4,6,4,9,15,6,15,35]).reshape((3,3))
q = np.array([5,7,6]).reshape((3,1))

def Newton(obj_grad, f, P, q, x0):
    repeat = True
    target_seq= [f(x0),]
    x = x0
    x1 = x - np.linalg.inv(P)@obj_grad(x)
    count = 0
    while f(x) != f(x1):
        count+=1
        x = x1
        x1 = x - np.linalg.inv(P)@obj_grad(x)
    print('Iter: ', count)
    print('X_opt: ', x1)
    print('f(x_opt): ',f(x1))
```

```python
Newton(obj_grad, f, P, q, x0)
```

### Davidon-Fletcher-Powell

```python
dfp_results = minimize(f, x0, method='Powell', options={'disp':True, 'return_all':True})
```

```python
dfp = plot_results(dfp_results, 'Davidon-Fletcher-Powell Results')
```

### Broyden-Fletcher-Goldfarb-Shanno

```python
bfgs_results = minimize(f, x0, method='BFGS', options={'disp':True, 'return_all':True})
```

```python
bfgs = plot_results(bfgs_results, 'BFGS Results')
```
